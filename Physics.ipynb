```python
# Importing modulels

import matplotlib.pyplot as plt # Ploting
import seaborn as sns           # Theme
import math                     # math.exp(arg)

def draw(title, function, n):
    '''To draw graph'''
    plt.plot(range(n), function(5), 'go-.' , linewidth=2.0)

    sns.set_style("darkgrid")  # darkgrid, whitegrid, dark, white, ticks
    plt.title(title)
    y, x = title.split('vs')
    plt.xlabel(x.strip())
    plt.ylabel(y.strip())

    plt.show()
```

```python
# Function as formulae

## Charge Vs Time
q_vs_t = lambda n: [round(1-(math.exp(-i)), 3) for i in range(1, n+1)]  # [0.632, 0.865, 0.95, 0.982, 0.993]
## Capacitance voltage vs time 
Vc_vs_t = lambda n: [round(1-(math.exp(-i)), 3) for i in range(1, n+1)]  # [0.632, 0.865, 0.95, 0.982, 0.993
## Current vs constant time capacitance
i_vs_Tc =  lambda n: [round((math.exp(-i)), 3) for i in range(1, n+1)]   # [0.368, 0.135, 0.05, 0.018, 0.007]
## Resistance voltage vs time
Vr_vs_t  = lambda n: [round((math.exp(-i)), 3) for i in range(1, n+1)]   # [0.368, 0.135, 0.05, 0.018, 0.007]
```

```python
## Charge Vs Time
draw(function=q_vs_t, n =5, title='Charge(q) vs Time(t)')
```

```python
## Capacitance voltage vs time 
draw(function=Vc_vs_t, n =5, title='Capacitance Voltage(Vc) vs Time(t)')
```

```python
## Current vs constant time capacitance
draw(function=i_vs_Tc, n =5, title='Current(I) vs Constant time capacitance(Tc)')
```

```python
## Resistance voltage vs time
draw(function=Vr_vs_t, n =5, title='Resistance voltage(Vr) vs time(t)')
```
